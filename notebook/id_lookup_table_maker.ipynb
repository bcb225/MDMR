```python
#fMRIPrep 1차 완료된 사람만 추출
directory_path = '/mnt/NAS2/data/SAD_gangnam_resting/fMRIPrep'
path = Path(directory_path)
fmri_code = []
# Iterate through the items in the directory
for item in path.iterdir():
    # Check if the item name starts with 'sub-' and does not end with '.html'
    if item.name.startswith('sub-') and not item.name.endswith('.html'):
        fmri_code.append(item.name[4:])
```

```python
#fMRIprep 1차 완료된 환자 code를 이용하여 screening #, enrollment #, HAID ID, 이름을 연결하는 파일을 만드는 함수
def generate_fmri_codes(id_table, fmri_code_list):
    # Create a copy of the dataframe to avoid modifying the original
    df = id_table.copy()
    
    # Define a function to generate fmri codes
    def generate_code(row):
        prefix = 's' if row['Enrollment #'].startswith('EXP') else 'c'
        code = f"{prefix}{int(row['Subject #']):04d}"
        return code
    
    # Apply the function to each row
    df['fmri_code'] = df.apply(generate_code, axis=1)
    
    # Filter the dataframe to include only rows where the fmri_code is in fmri_code_list
    df_filtered = df[df['fmri_code'].isin(fmri_code_list)]
    
    # Select relevant columns and rearrange them
    result = df_filtered[['fmri_code', 'Screening #', 'Enrollment #', 'HAID ID', '이름']]
    
    return result

result_df = generate_fmri_codes(id_table, fmri_code)
```
